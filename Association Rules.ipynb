Load the dataset

```python
import pandas as pd

df = pd.read_excel("/content/Online retail.xlsx", header=None)

df.head()
```

Check missing values

```python
df.isnull().sum()
```

Remove duplicate transactions

```python
df = df.drop_duplicates()
```

Convert rows into transactions

```python
transactions = []

for i in range(len(df)):
    transaction = []

    for item in df.iloc[i]:
        if pd.notna(item):
            transaction.extend(str(item).split(','))

    transactions.append(transaction)

transactions[:5]
```

Install Mlxtend

```python
!pip install mlxtend
```

Transaction Encoder

```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()

te_array = te.fit(transactions).transform(transactions)

basket = pd.DataFrame(te_array, columns=te.columns_)

basket.head()
```

Apply Apriori

```python
from mlxtend.frequent_patterns import apriori

frequent_items = apriori(
    basket,
    min_support=0.02,
    use_colnames=True
)

frequent_items.head()
```

Generate Association Rules

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(
    frequent_items,
    metric="lift",
    min_threshold=1
)

rules.head()
```

Strong Rules

```python
rules = rules.sort_values(by="lift", ascending=False)

rules.head(20)
```

1. What is Lift and why is it important in Association Rules?

Ans;

Lift measures how much more likely two items are purchased together compared to if they were independent.

Formula:

Lift = Confidence(A → B) / Support(B)
Lift > 1 → Positive association
Lift = 1 → No association
Lift < 1 → Negative association

Lift is important because it identifies meaningful relationships beyond simple co-occurrence.

2. What are Support and Confidence? How do you calculate them?

Support

Support measures how frequently an itemset appears in the dataset.

Formula:

Support(A) = Number of transactions containing A / Total transactions

Confidence

Confidence measures the probability that item B is purchased when item A is purchased.

Formula:

Confidence(A → B) = Support(A ∪ B) / Support(A)

Higher confidence indicates stronger association rules.

3. What are some limitations or challenges of Association Rule Mining?

Ans:

Generates a large number of rules, many of which may not be useful.
Choosing appropriate support and confidence thresholds can be difficult.
Computational cost increases with larger datasets.
Association rules identify relationships but do not imply causation.
